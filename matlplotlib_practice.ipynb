  ##                  BASICS of Matplotlib

```python
import matplotlib.pyplot  as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
plt.title('Weather',color = 'Blue')
plt.xlabel('days',color = 'yellow')
plt.ylabel('Temperature',color = 'red')
days = [1,2,3,4,5]
temp = [10,15,14,8,20]
temp_max = [20,32,18,29,34]
temp_min = [2,3,1,2.4,3.4]
plt.plot(days,temp,color = 'blue',linewidth = 2,linestyle  = 'dashed',label = 'Avg')
plt.plot(days,temp_max,color = 'red',linewidth = 2,linestyle  = 'dotted',label = 'Max')
plt.plot(days,temp_min,color = 'green',linewidth = 4,label = 'Min')
plt.legend(loc = 'upper right',shadow = 'True',fontsize = 'small')
plt.grid()
```

```python
plt.title('Weather',color = 'Blue')
plt.xlabel('days',color = 'yellow')
plt.ylabel('Temperature',color = 'red')
days = [1,2,3,4,5]
temp = [10,15,14,8,20]
temp_max = [20,32,18,29,34]
temp_min = [2,3,1,2.4,3.4]
plt.plot(days,temp,"--g",marker = 'D',markersize =10,alpha = 0.2) # alpha --transparency
```

## Bar Chart

```python
company = ['amazon','Google','Microsoft','Facebook']
revenues = [200,300,150,400]
profit = [100,140,30,200]
#bar chart
xpos = np.arange(len(company))
plt.xticks(xpos,company)
plt.ylabel("revenues")
plt.bar(xpos-0.2,revenues,width = 0.4,color = 'green',label = 'Revenues')
plt.bar(xpos+0.2,profit,width=0.4,color = 'blue',label = 'Profits')
plt.legend()
plt.title("Revenues of the tech company",color = 'yellow')
```

```python
company = ['amazon','Google','Microsoft','Facebook']
revenues = [200,300,150,400]
profit = [100,140,30,200]
#bar chart
xpos = np.arange(len(company))
plt.yticks(xpos,company)
plt.ylabel("revenues")
plt.barh(xpos-0.2,revenues,color = 'green',label = 'Revenues')
plt.barh(xpos+0.2,profit,color = 'red',label = 'Profits')
plt.legend()
plt.title("Revenues of the tech company")  #barh : horizontal graph,need to change xticks to yticks
```

## Histograms

```python
blood_sugar = [80,75,90,120,210,70,40,140,200,130,170]
plt.hist(blood_sugar,bins =[40,70,110,125,140,210],rwidth = 0.95,color = 'purple') #by default bins is preset.
#rwidth : separate the bars
```

```python
blood_sugar_men = [80,75,90,120,210,70,40,140,200,130,170]
blood_sugar_women = [70,78,93,90,160,70,50,153,119,136,158]
plt.hist([blood_sugar_men,blood_sugar_women],bins =[40,70,110,125,140,210],rwidth = 0.95,color = ['purple','orange'],label=['men_sugar','women_sugar']) #by default bins is preset.
#rwidth : separate the bars
plt.xlabel("Sugar range",color = 'red')
plt.ylabel("Total no.of patients",color = 'red')
plt.legend()
plt.title("Blood sugar levels of men and women",color = 'red')
```

```python
blood_sugar_men = [80,75,90,120,210,70,40,140,200,130,170]
blood_sugar_women = [70,78,93,90,160,70,50,153,119,136,158]
plt.hist([blood_sugar_men,blood_sugar_women],bins =[40,70,110,125,140,210],rwidth = 0.95,color = ['purple','orange'],label=['men_sugar','women_sugar'],orientation='horizontal') #by default bins is preset.
#rwidth : separate the bars
plt.xlabel("Sugar range",color = 'red')
plt.ylabel("Total no.of patients",color = 'red')
plt.legend()
plt.title("Blood sugar levels of men and women",color = 'red')
```

## Pie Chart

```python
scores = [20,40,102,65,12,47]
batsmen = ["Sehwag","Tendulkar","Kohli","Yuvraj","Dhoni","Pandya"]
```

```python
plt.pie(scores,labels=batsmen,radius=1.5,autopct="%0.2f%%",shadow=True,explode= [0,0.2,0,0,0,0],startangle=45)
plt.savefig("piechart.png",bbox_inches = 'tight',pad_inches=2,transparent=True)
```
